# [BONUS] Plot X, Y and Best Fit Line

**Write a function to calculate best fit.**

![alt text](https://i.imgur.com/B43vOV0.jpg)[1]

[1]: https://www.mathsisfun.com/data/least-squares-regression.html "Least Squares Regression"

````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# formulas
# y = mx + b
# m =  ( N Σ(xy) − Σx Σy ) / ( N Σ(x2) − (Σx)sq )
# b =  ( Σy − m Σx ) / N
# N is the no. of points = 5

# solve for a and b
def best_fit(X, Y):
    l = len(X)
    sumX = 0
    sumY = 0
    sumXY = 0
    sumXX = 0
    for i in range(l):
        sumX += X[i]
        sumY += Y[i]
        sumXY += X[i] * Y[i]
        sumXX += X[i] * X[i]
    
    m = ((l*sumXY)-(sumX*sumY)) / ((l*sumXX)-(sumX*sumX))
    b = ( sumY-m*sumX ) / l
    return m, b

# solution
m, b = best_fit(X, Y)
#best fit line:
#y = 0.80 + 0.92x

# Create Z list based on "y = mx + b"
Z = [0]*len(X)
for i in range(len(X)):
    Z[i] = m*X[i] + b

# To Do: plot points and fit line
plt.scatter(X,Y)
plt.plot(X, Z)
````

# [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem
2. scipy.stats.t.ppf

````python
# Null Hypothesis = We assume b different from A.
# We ran random chance for difference(A,B)
# If difference(A,B) > Confidence Interval, we reject Null Hypothesis.

import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):   
    # To Do - Calculate m, h
    a = 1.0*np.array(data)
    n = len(a)
    m = np.mean(a)
    se = stats.sem(a)
    h = se * stats.t._ppf((1+confidence)/2., n-1)
    
    return m, m-h, m+h

data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
print(mean_confidence_interval(data))
````
